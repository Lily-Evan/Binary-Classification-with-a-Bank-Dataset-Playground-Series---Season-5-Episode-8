```python
import pandas as pd
import numpy as np

# Φορτώνουμε submissions
sub_llm = pd.read_csv('/kaggle/input/submission-97338/submission_097338.csv')[['id','y']]
sub_ml  = pd.read_csv('/kaggle/input/submission-97763/submission_97763.csv')[['id','y']]

# Merge στο id
merged_df = pd.merge(sub_ml, sub_llm, on='id', suffixes=('_ML', '_LLM'))

# Helper: prob -> logit, logit -> prob
def prob_to_logit(p):
    eps = 1e-15
    p = np.clip(p, eps, 1-eps)
    return np.log(p/(1-p))

def logit_to_prob(l):
    return 1/(1+np.exp(-l))

# Μετατρέπουμε σε logits
logit_ml  = prob_to_logit(merged_df['y_ML'].values)
logit_llm = prob_to_logit(merged_df['y_LLM'].values)

# Grid search weights στο logit space
for w in np.linspace(0,1,21):  # 0.0, 0.05, ..., 1.0
    blended_logit = w * logit_ml + (1-w) * logit_llm
    blended_prob  = logit_to_prob(blended_logit)
    
    submission = pd.DataFrame({
        "id": merged_df['id'],
        "y": blended_prob
    })
    submission.to_csv(f"/kaggle/working/submission_logitblend_{w:.2f}.csv", index=False)
    print(f"Saved logit-blended submission with weight {w:.2f}")
```
